### Exercic√≠os

Implementar um script em linguagem R para calcular a
m√©dia de hp dos carros com 8 cilindros baseado no
dataset mtcars. Utilizar o conhecimento adquirido at√© o
momento. Por exemplo, evitar fun√ß√µes de filtro e m√©dia.

````r
data(mtcars)
head(mtcars)
````

````r
carros_8cil <- mtcars[mtcars$cyl == 8,]
valores_hp <- carros_8cil$hp
soma_hp <- 0

for (valor in valores_hp) {
  soma_hp <- soma_hp + valor
}

# Contar quantos carros t√™m 8 cilindros
n_carros <- 0
for (i in 1:length(valores_hp)) {
  n_carros <- n_carros + 1
}

# Calcular a m√©dia (soma / quantidade)
media_hp <- soma_hp / n_carros

# Exibir o resultado
print(paste("M√©dia de hp (8 cilindros):", media_hp))
````

- Verifique se x √© par ou √≠mpar usando %%

  - Imprima ‚ÄúPar‚Äù ou ‚Äú√çmpar‚Äù


üìå Crie um script que:

* Defina tr√™s vari√°veis: n1, n2 e n3

Calcule a m√©dia das tr√™s

- Use if e else para:

- Imprimir ‚ÄúAprovado‚Äù se a m√©dia for maior ou igual a 7

- Caso contr√°rio, imprimir ‚ÄúReprovado‚Äù

````r
n1 <- as.numeric(readline(prompt="Digite sua nota n1: "))
n2 <- as.numeric(readline(prompt="Digite sua nota n2: "))
n3 <- as.numeric(readline(prompt="Digite sua nota n3: "))

media = (n1 + n2 + n3)/3

if(media >= 7) {
  print("Aprovado")
} else {
  print("Reprovado")
}
````

üìå Use um la√ßo for para:

Imprimir a tabuada do n√∫mero 5 de 1 at√© 10

Exemplo de sa√≠da:
5 x 1 = 5
5 x 2 = 10

````r
n <- as.integer(readline(prompt="Digite um n√∫mero inteiro para saber sua tabuada at√© 10: "))

for(i in 1:10) {
  resultado <- n * i
  print(paste(n, "x", i, "=", resultado))
}
````

üìå Escreva um c√≥digo que:

Use um la√ßo for para somar todos os n√∫meros pares de 1 a 20

Ao final, imprima o total da soma

````r
soma <- 0

for (i in 1:20) {
  if (i %% 2 == 0) {
    soma <- soma + i
  }
}

print(soma)
````
