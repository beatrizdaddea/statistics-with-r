### Exercicíos

Implementar um script em linguagem R para calcular a
média de hp dos carros com 8 cilindros baseado no
dataset mtcars. Utilizar o conhecimento adquirido até o
momento. Por exemplo, evitar funções de filtro e média.

````r
data(mtcars)
head(mtcars)
````

````r
carros_8cil <- mtcars[mtcars$cyl == 8,]
valores_hp <- carros_8cil$hp
soma_hp <- 0

for (valor in valores_hp) {
  soma_hp <- soma_hp + valor
}

# Contar quantos carros têm 8 cilindros
n_carros <- 0
for (i in 1:length(valores_hp)) {
  n_carros <- n_carros + 1
}

# Calcular a média (soma / quantidade)
media_hp <- soma_hp / n_carros

# Exibir o resultado
print(paste("Média de hp (8 cilindros):", media_hp))
````

 ## Exercícios de R – Conceitos Iniciais (Para você escrever o código)

📌 Escreva um código em R que:

- Leia um número x

- Verifique se x é par ou ímpar usando %%

  - Imprima “Par” ou “Ímpar”

````r
x <- as.integer(readline(prompt="Digite um número inteiro: "))

if (x %% 2 == 0) {
  print("Par")
} else {
  print("Ímpar")
}
````


📌 Crie um script que:

* Defina três variáveis: n1, n2 e n3

Calcule a média das três

- Use if e else para:

- Imprimir “Aprovado” se a média for maior ou igual a 7

- Caso contrário, imprimir “Reprovado”

````r
n1 <- as.numeric(readline(prompt="Digite sua nota n1: "))
n2 <- as.numeric(readline(prompt="Digite sua nota n2: "))
n3 <- as.numeric(readline(prompt="Digite sua nota n3: "))

media = (n1 + n2 + n3)/3

if(media >= 7) {
  print("Aprovado")
} else {
  print("Reprovado")
}
````

📌 Use um laço for para:

Imprimir a tabuada do número 5 de 1 até 10

Exemplo de saída:
5 x 1 = 5
5 x 2 = 10

````r
n <- as.integer(readline(prompt="Digite um número inteiro para saber sua tabuada até 10: "))

for(i in 1:10) {
  resultado <- n * i
  print(paste(n, "x", i, "=", resultado))
}
````

📌 Escreva um código que:

Use um laço for para somar todos os números pares de 1 a 20

Ao final, imprima o total da soma

````r
soma <- 0

for (i in 1:20) {
  if (i %% 2 == 0) {
    soma <- soma + i
  }
}

print(soma)
````

##  Questões de R: Conceitos Iniciais, Condicionais e Laços

**10 questões práticas de código em R** para revisar os **conceitos iniciais**, incluindo **operadores lógicos, estruturas condicionais (`if`, `else`) e laços (`for`, `while`)**:

---

### 📘 **Questões de R: Conceitos Iniciais, Condicionais e Laços**

---

#### **1. Operadores Lógicos**

```r
# O que será impresso no console ao executar o código abaixo?

x <- 5
y <- 10

(x > 3) & (y < 15)
```

**a)** **TRUE**

**b)** FALSE

**c)** NA

**d)** Erro de sintaxe

---

#### **2. Operadores de Comparação**

```r
# Qual será o valor de z?

a <- 7
b <- 12
z <- (a == b) | (a < b)
```

**a)** **TRUE**

**b)** FALSE

**c)** NA

**d)** NULL

---

#### **3. Condicional Simples**

```r
x <- 20

if (x %% 2 == 0) {
  print("Par")
}
```

O que será impresso?

**a)** **Par**

**b)** Ímpar

**c)** Erro

**d)** NULL

---

#### **4. Condicional Composta**

```r
nota <- 6.5

if (nota >= 7) {
  print("Aprovado")
} else {
  print("Reprovado")
}
```

Se `nota <- 6.5`, qual será a saída?

Resposta: **Reprovado**
---

#### **5. Condicional Aninhada**

```r
idade <- 16

if (idade >= 18) {
  print("Maior de idade")
} else if (idade >= 12) {
  print("Adolescente")
} else {
  print("Criança")
}
```

Qual valor será impresso?

Resposta: **Adolescente**
---

#### **6. Estrutura de Repetição `for`**

```r
# O que será impresso?

for (i in 1:3) {
  print(i^2)
}
```

**a)** **1 4 9**

**b)** 1 2 3

**c)** 1 3 9

**d)** Erro

---

#### **7. Estrutura `while`**

```r
i <- 1
while (i <= 3) {
  print(i)
  i <- i + 1
}
```

Qual é a saída desse laço?

Resposta: **1 2 3**
---

#### **8. Uso de `next`**

```r
for (i in 1:5) {
  if (i == 3) {
    next
  }
  print(i)
}
```

Qual número será **ignorado** na impressão?

Resposta: **o número ignorado é 3**
---

#### **9. Uso de `break`**

```r
for (i in 1:5) {
  if (i == 4) {
    break
  }
  print(i)
}
```

Quantos valores serão impressos?

Resposta: ** 3 valores impressos**
---

#### **10. Condicional com Vetor**

```r
x <- c(10, 15, 20)

x > 12
```

Qual é a saída?

**a)** TRUE

**b)** FALSE

**c)** **FALSE TRUE TRUE**

**d)** TRUE TRUE TRUE
