# Função de Distribuição Normal e Binomial

## Densidade de probabilidade em um determinado ponto

### Exemplo 1: funções para distribuição normal

- N(média,desvio- padrão)
- Média: 100
- Variância: 64
- N(100,8)


```r
x <- seq(70, 130, len = 100) # gera uma sequência de número
fx <- dnorm(x, 100, 8) # é a função de distribuição normal
fx
plot(x, fx, type = "l") # define o tipo de gráfico como linha ("l" de line)
Fx <- pnorm(x, 100, 8)
Fx
plot(x, Fx, type = "l")
```

### Exemplo 2:  funções para distribuição binomial
- n: 10
- p: 0.35

```r
x <- 1:10
fx <- dbinom(x, 10, 0.35) # é a função de distribuição binomial
fx
plot(x, fx, type = "h") # gráfico histograma
Fx <- pbinom(x, 10, 0.35)
Fx
plot(x, Fx, type = "s") # gráfico de escada
```

A duração de um processo em uma linha de montagem para fabricar um componente é em média de 230 minutos, com desvio-padrão de 15 minutos.
Verificar a probabilidade de um componente ser produzido em menos de 240,5 minutos.

**Observação: os dados de tempo possuem uma distribuição normal**

```r
#1º passo: levantamento dos dados da situação
#Valor: 240,5 minutos
#Média: 230 minutos
#Desvio-padrão: 15 minutos

#2º passo: atribuição dos dados para variáveis (implementação)
valor <- 240.5
media <- 230
desvio_padrao <- 15

#3º passo: cálculo utilizando a equação
z <- (valor - media)/desvio_padrao
z

#4º passo: obtenção de valor da tabela da normal reduzida
#0,2580 considerando z = 0,70

#5º passo: cálculo da probabilidade P
#P(Tempo <= 240,5) = 0.5 (área da função calculada com valores antes da média) + 0.2580 (área calculada com valores após a média)
#P(Tempo <= 240,5) = 0,5 + 0,2580 = 0,7580
```

## Intervalo de Confiança

```r
#1º passo: extração dos dados e do objetivo
#Calcular o intervalo de confiança com base:
#- média amostral;
#- variância populacional.
#Nível de confiança: 0,95
#Pressuposto: distribuição normal

#Desvio-padrão populacional: 5 horas
#Tamanho da amostra (n): 100 peças
#Média amostral: 500 horas

#2º passo: criação de variáveis e atribuição de valores
desvio_padrao <- 5
n <- 100
X <- 500

conf_level <- 0.95
alpha <- 1 - conf_level
z_critical <- qnorm(1 - alpha/2) # quantis da distribuição normal

z_critical

#3º passo: criação de variável e atribuição do valor
#do Z crítico
#Uso da tabela de distribuição, área de 0,475
Zcritico <- 1.96

#4º passo: aplicação da equação para cálculo do
#intervalo de confiança
#Intervalo de confiança - soma
ICSoma <- X + Zcritico * desvio_padrao/sqrt(n)

#Intervalo de confiança - subtração
ICSubtracao <- X - Zcritico * desvio_padrao/sqrt(n)

#5º passo: apresentação do resultado
cat("Intervalo de Confiança (95%): [", ICSubtracao
, ", ", ICSoma, "]\n")
```

### Considere uma situação em que a variância populacional é de 25 horas

```r
#Nível de confiança: 0,95
#Pressuposto: distribuição normal

#Desvio-padrão populacional: 5 horas
# Variância populacional = 𝜎^2= 25 horas
# Sendo assim o 𝜎= 5 horas

#Tamanho da amostra (n): 100 peças
#Média amostral: 500 horas

#2º passo: criação de variáveis e atribuição de valores
desvio_padrao <- 5
n <- 100
X <- 500

#3º passo: criação de variável e atribuição do valor
#do Z crítico
#Uso da tabela de distribuição, área de 0,475
Zcritico <- 1.96

#4º passo: aplicação da equação para cálculo do
#intervalo de confiança
#Intervalo de confiança - soma
ICSoma <- X + Zcritico * desvio_padrao/sqrt(n)

#Intervalo de confiança - subtração
ICSubtracao <- X - Zcritico * desvio_padrao/sqrt(n)

#5º passo: apresentação do resultado
cat("Intervalo de Confiança (95%): [", ICSubtracao
, ", ", ICSoma, "]\n")
```

### Considere uma situação em que o nível de confiança seja de 90%

```r
#Nível de confiança: 0,90
#Pressuposto: distribuição normal

#Desvio-padrão populacional: 5 horas
#Tamanho da amostra (n): 100 peças
#Média amostral: 500 horas

#2º passo: criação de variáveis e atribuição de valores
desvio_padrao <- 5
n <- 100
X <- 500

#3º passo: criação de variável e atribuição do valor
#do Z crítico
#Uso da tabela de distribuição, área de 0,450
Zcritico <- 1.64

#4º passo: aplicação da equação para cálculo do
#intervalo de confiança
#Intervalo de confiança - soma
ICSoma <- X + Zcritico * desvio_padrao/sqrt(n)

#Intervalo de confiança - subtração
ICSubtracao <- X - Zcritico * desvio_padrao/sqrt(n)

#5º passo: apresentação do resultado
cat("Intervalo de Confiança (90%): [", ICSubtracao
, ", ", ICSoma, "]\n")
```

### Considere uma situação em que o tamanho da população seja de 1000 peças

```r
# Dados do problema
N <- 1000            # Tamanho da população
n <- 100             # Tamanho da amostra
X <- 500             # Média amostral
desvio_padrao <- 5   # Desvio padrão populacional
confianca <- 0.95    # Nível de confiança (95%)

# 1. Calcular o valor crítico Z para 95% de confiança
Zcritico <- qnorm(1 - (1 - confianca)/2)  # Z = 1.96

# 2. Calcular o Erro Padrão (EP) com Fator de Correção para População Finita (FCP)
# Fórmula: EP_ajustado = (σ/√n) * √((N - n)/(N - 1))
EP_ajustado <- (desvio_padrao / sqrt(n)) * sqrt((N - n)/(N - 1))

# 3. Calcular a Margem de Erro (ME)
ME <- Zcritico * EP_ajustado

# 4. Calcular os limites do Intervalo de Confiança (IC)
ICSoma <- X + ME
ICSubtracao <- X - ME

# 5. Apresentar o resultado
cat("Intervalo de Confiança (95%) com FCP: [", round(ICSubtracao, 2), ", ", round(ICSoma, 2), "]\n")
```
